Estructuración estricta de como debe de entregar la información, definiendo, los campos que se desean en la salida del prompt.
Ejemplo: "Genera un diccionario de datos para el siguiente archivo CSV, incluyendo los campos: Nombre del campo, Tipo de dato, Descripción, Restricciones y Relaciones."

*   **Prompting Iterativo y Refinamiento:**

Realizaremos múltiples iteraciones de prompts, refinando las preguntas y ajustando los parámetros para obtener los resultados deseados.

````python
import os
import pandas as pd
import openai
from dotenv import load_dotenv

# =============================
# Configuración y preparación
# =============================

load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")

CARPETA_ARCHIVOS = "./archivos_csv"
CARPETA_RESPUESTAS = "./respuestas_chatgpt"

PREGUNTAS_CABLE = [
    "Analiza el archivo. Extrae los nombres de las columnas, determina el tipo de datos de cada columna (texto, número, fecha) y proporciona una breve descripción del contenido de cada columna. Presenta la información en formato de tabla con las columnas 'Nombre de Columna', 'Tipo de Dato' y 'Descripción'.",
    "Para el archivo, genera una lista de todos los campos con sus correspondientes tipos de dato. Quiero que la salida esté en formato de lista, con el siguiente formato 'campo : tipo de dato'.",
    "Crea un diccionario de datos completo para el archivo. Incluye los campos: 'Nombre del Campo', 'Tipo de Dato', 'Descripción', 'Restricciones' (por ejemplo, valores permitidos, longitud máxima) y 'Relaciones' (si aplica)."
]

# =============================
# Funciones principales
# =============================

def obtener_muestra_de_archivo(ruta, max_filas=20):
    try:
        df = pd.read_csv(ruta)
        return df.head(max_filas).to_csv(index=False)
    except Exception as e:
        return f"No se pudo leer el archivo correctamente: {str(e)}"

def consultar_chatgpt_con_archivo(nombre_archivo, contenido_csv, pregunta):
    prompt = f"""Este es un archivo CSV llamado '{nombre_archivo}'. A continuación tienes una muestra de su contenido:

```csv
{contenido_csv}
````

````python
respuesta = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "Eres un experto en análisis de datos y estructuras tabulares."},
        {"role": "user", "content": prompt}
    ]
)
return respuesta['choices'][0]['message']['content']
````
